```python
import pandas as pd
import numpy as np
import nltk
from nltk import pos_tag, word_tokenize
import re
import csv
```

```python
data = pd.read_csv('/Users/naksh/Documents/Sem_3/NLP/Project/inference_results.csv')
```

```python
answer = data["answer"]
generated_answer = data["generated_answer"]
```

```python
from nltk.translate.bleu_score import sentence_bleu
from nltk import word_tokenize

def calculate_bleu(candidate, reference):
    '''
    candidate, reference: generated and ground-truth sentences
    '''
    reference = word_tokenize(reference)
    candidate = word_tokenize(candidate)
    score = sentence_bleu(reference, candidate)
    return score

scores = []
    
    
for i in range(len(answer)):
    score = calculate_bleu(answer[i], generated_answer[i])
    scores.append(round(score,2))
print(scores)
```

```python
def calculate_rouge_score(reference, candidate, n=1):
    reference = [reference.split()]  # Convert reference sentence to a list of words
    candidate = candidate.split()  # Convert candidate sentence to a list of words

    # Calculate ROUGE-N score
    rouge_score = sentence_bleu(reference, candidate, weights=(1, 0, 0, 0)[:n])

    return rouge_score
scores = []
    
    
for i in range(len(answer)):
    score = calculate_rouge_score(answer[i], generated_answer[i])
    scores.append(round(score,2))
avg = sum(scores)/len(scores)
print("Rouge Score:",avg*100)
```

```python
def extract_object(sentence):
    tokens = word_tokenize(sentence)
    tagged_tokens = pos_tag(tokens)

    for token, pos in tagged_tokens:
        if pos in ['NN', 'NNS']:
            return token
    return None

def extract_positions(sentence):
    
    pattern = r'\b(topLeft|topCenter|topRight|middleRight|middleCenter|middleLeft|bottomLeft|bottomCenter|middleRight)\b'

    # Search for the pattern in the sentence
    match = re.search(pattern, sentence)

    # If a match is found, return the matched position
    if match:
        return match.group(1)
    else:
        return None
```
